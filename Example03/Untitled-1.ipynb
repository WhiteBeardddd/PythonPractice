```python
import numpy as np
import matplotlib.pyplot as plt

# Moving Average Filter
def movingAverageFilter(x, L):
    y = np.zeros_like(x)
    for n in range(len(x)):
        if n < L:
            y[n] = np.mean(x[:n+1])
        else:
            y[n] = np.mean(x[n-L+1:n+1])
    return y

# Parameters
fs = 8000       # Sampling frequency
f = 440         # Signal frequency
dur = 0.5       # Duration in seconds
L = 5           # Window length for moving average
point = 200     # For plotting only first samples

# Time index
N = int(fs * dur)
n = np.arange(N)

# Original sine wave
x = np.sin(2 * np.pi * f * n / fs)

# Add random noise
noise_strength = 0.5
x_noisy = x + noise_strength * np.random.randn(N)

# Apply moving average
y = movingAverageFilter(x_noisy, L)

# Plot results
plt.figure(figsize=(12, 6))

plt.subplot(3,1,1)
plt.plot(n[:point], x[:point])
plt.title("Original 440 Hz Sine Wave")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)

plt.subplot(3,1,2)
plt.plot(n[:point], x_noisy[:point])
plt.title("Noisy Signal")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)

plt.subplot(3,1,3)
plt.plot(n[:point], y[:point])
plt.title(f"Smoothed Signal with Moving Average (L={L})")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)

plt.tight_layout()
plt.show()
```
